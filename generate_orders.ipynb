```python
import random
insert_queries = []

for n_dt, dt in enumerate(pd.date_range('1-1-2018', '29-4-2022')):

    n_orders = np.random.randint(int(n_dt * .01 + (dt.month +1)* .2 + 5), int(n_dt * .015 + (dt.month +2)* .3 + 10))

    for n_ord in range(n_orders):

        customer_id = random.choice(cust_ids)

        cust_city = [cust_cities[i] for i in range(len(cust_ids)) if cust_ids[i] == customer_id ][0]

        n_cust_ord = np.random.randint(1, int(customer_id/40) + 3)

        for nnn in range(n_cust_ord):

            prod_cat_options = random.choice(all_product_categories)

            prod_names = (prod_table['product_name'][prod_table['product_category'] == prod_cat_options]).tolist()
            prod_ids = (prod_table['product_id'][prod_table['product_category'] == prod_cat_options]).tolist()
#             prod_prices = [i[2] for i in data]
#             seller_ids =  [i[3] for i in data]
            
            
            prod_opt = random.choice(prod_names)
            prod_id = [prod_ids[i] for i in range(len(prod_ids)) if prod_names[i] == prod_opt ][0]
#             price = [prod_prices[i] for i in range(len(prod_prices)) if prod_names[i] == prod_opt ][0]
#             seller_id = [seller_ids[i] for i in range(len(seller_ids)) if prod_names[i] == prod_opt ][0]
            
            
            Quantity = np.random.randint(1,11)
            payment_type = random.choice(["debit_card", "credit_card", "food_stamp", "cash" , "gift_card"])

            order_id = int(str(dt.year)[-2:] + ('0' + str(dt.day))[-2:] + ('0' + str(dt.month))[-2:] + ('000' + str(customer_id))[-3:])


            vals_1 = ', '.join([str(order_id), str(customer_id), "'{}'".format(str(dt.date())), "'{}'".format(payment_type)])
            vals_2 = ', '.join([str(order_id), str(prod_id), str(Quantity)])

            insert_query1 = """INSERT into orders (order_id, customer_id, purchase_time, payment_type) VALUES (""" + vals_1 + ") ON CONFLICT DO NOTHING"
            insert_query2 = """INSERT into order_items (order_id ,product_id, quantity) VALUES (""" + vals_2 + ")"

            if nnn == 0:
                insert_queries.append(insert_query1)
            
            insert_queries.append(insert_query2)     
    print(dt)
```
